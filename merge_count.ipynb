```python
path = r"C:/Users/dharm/Documents/Biostate/count_files/featurecounts_output2/"
```

```python
# Initialize an empty DataFrame to store merged data
merged_df = pd.DataFrame()

# Loop through all files in the specified directory
for filename in os.listdir(path):
    if filename.endswith(".txt"):
        # Read the file
        file_path = os.path.join(path, filename)
        df = pd.read_csv(file_path, sep="\t", skiprows =1)  # Adjust the delimiter if needed

        # Extract the gene_name and last column (sample counts)
        sample_name = filename.split('_no')[0]  # Keep the part before '_no_duplicates_counts'
        df = df[['Geneid', df.columns[-1]]]  # Select 'gene_name' and last column
        df.columns = ['Geneid', sample_name]  # Rename columns

        # Merge with the existing DataFrame on 'gene_name'
        if merged_df.empty:
            merged_df = df
        else:
            merged_df = pd.merge(merged_df, df, on='Geneid', how='outer')

# Save the merged DataFrame to a file
output_path = os.path.join(path, "merged_counts.txt")
merged_df.to_csv(output_path, sep="\t", index=False)
#print(f"Merged count file saved to {output_path}")
```
